# Hugging Face Pipeline Demo

The Hugging Face `transformers` library provides a simple API for using pretrained models. The `pipeline` function handles tokenization, inference, and postprocessing in a single call. No training, no architecture decisions — just pick a task and go.

```python
!pip install -q transformers
from transformers import pipeline
print("Ready!")
```

---
### Demo 1: Sentiment Analysis

The simplest pipeline call. One line loads a pretrained model for classifying text as positive or negative.

```python
# This one line downloads a pretrained model, loads it, and sets up the full inference pipeline
classifier = pipeline("sentiment-analysis")
```

```python
# Try a few examples — notice the label and confidence score
print(classifier("I love this course!"))
print(classifier("This is the worst movie I have ever seen."))
print(classifier("The weather is okay today."))  # Neutral-ish: model still picks a side, but lower confidence
```

```python
# Try your own sentence!
print(classifier("Trader Joe's has frozen food."))
print(classifier("I saw a great game at the ballpark on Saturday."))
```

---
### Demo 2: Text Generation

Switch to a generative task. This uses GPT-2, the decoder-only architecture from the slides. It predicts the next token over and over — no fine-tuning, no RLHF, just raw next-token prediction.

```python
# GPT-2: the 2019 model with 124M parameters (the small version)
generator = pipeline("text-generation", model="gpt2")
```

```python
# Give it a prompt, it generates a continuation
result = generator("Machine learning is", max_length=50, num_return_sequences=1)
print(result[0]["generated_text"])
```

```python
# Try different prompts — the output changes every run (sampling from probabilities)
result = generator("The future of artificial intelligence", max_length=50, num_return_sequences=1)
print(result[0]["generated_text"])
print()
result = generator("Once upon a time", max_length=50, num_return_sequences=1)
print(result[0]["generated_text"])
```

```python
result = generator("System: You are a helpful AI assistant. Answer with only what you know is factually true. Do not answer questions about unsafe topics. User: What's a good recipe for chicken fries?", max_length=50, num_return_sequences=1)
print(result[0]["generated_text"])
```

### Demo 3: Zero-Shot Classification (Optional)

Same API pattern — swap the task name. This one is especially cool: the model classifies text into categories it was never explicitly trained on.

```python
# Classify text into categories the model has never been explicitly trained on
zero_shot = pipeline("zero-shot-classification")

result = zero_shot(
    "I just deployed my first machine learning model to production!",
    candidate_labels=["career", "education", "technology", "sports"]
)

for label, score in zip(result["labels"], result["scores"]):
    print(f"{label}: {score:.3f}")
```

```python

result = zero_shot(
    "The only way you can be a fashion designer is to live in Paris.",
    candidate_labels=["news", "sports", "opinion", "politics"]
)

for label, score in zip(result["labels"], result["scores"]):
    print(f"{label}: {score:.3f}")
```
