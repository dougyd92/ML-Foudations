# Understanding the target variable

**The target variable is what we want to predict.**

In this dataset, it's `MedHouseVal` (median house value). Understanding its distribution
is crucial because it affects which models and transformations might work best.

### EXAMPLE 2.1: Visualizing Target Distribution

We'll use both histograms and box plots to understand the distribution.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
axes[0].hist(df['MedHouseVal'], bins=50, edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Median House Value ($100,000s)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Distribution of House Values')
axes[0].axvline(df['MedHouseVal'].mean(), color='red', linestyle='--', label='Mean')
axes[0].axvline(df['MedHouseVal'].median(), color='green', linestyle='--', label='Median')
axes[0].legend()

# Box plot
axes[1].boxplot(df['MedHouseVal'])
axes[1].set_ylabel('Median House Value ($100,000s)')
axes[1].set_title('Box Plot of House Values')
axes[1].set_xticklabels(['MedHouseVal'])

plt.tight_layout()
plt.show()
```

**Observations:**
- The distribution is right-skewed (tail extends to the right)
- There's a spike at the maximum value (5.0) - this might be capped data
- The box plot shows several outliers on the high end

### EXAMPLE 2.2: Key Statistics of Target Variable

Let's calculate important statistical properties.

```python
target_stats = {
    'Mean': df['MedHouseVal'].mean(),
    'Median': df['MedHouseVal'].median(),
    'Std Dev': df['MedHouseVal'].std(),
    'Min': df['MedHouseVal'].min(),
    'Max': df['MedHouseVal'].max(),
    'Range': df['MedHouseVal'].max() - df['MedHouseVal'].min(),
    'Skewness': df['MedHouseVal'].skew()
}

for stat, value in target_stats.items():
    print(f"{stat:12}: ${value * 100000:.0f}")
```

**Skewness interpretation:**
- Skewness > 0: Right-skewed (our case)
- Skewness = 0: Symmetric
- Skewness < 0: Left-skewed

# SECTION 3: EXPLORING RELATIONSHIPS BETWEEN VARIABLES

Understanding how features relate to the target variable helps us:
- Identify the most predictive features
- Discover patterns and trends
- Guide feature selection and engineering

### EXAMPLE 3.1: Visualizing Feature-Target Relationships

Scatter plots are excellent for seeing relationships between continuous variables.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# MedInc vs MedHouseVal
axes[0, 0].scatter(df['MedInc'], df['MedHouseVal'], alpha=0.3, s=10)
axes[0, 0].set_xlabel('Median Income ($10,000s)')
axes[0, 0].set_ylabel('Median House Value ($100,000s)')
axes[0, 0].set_title('Income vs House Value')

# AveRooms vs MedHouseVal
axes[0, 1].scatter(df['AveRooms'], df['MedHouseVal'], alpha=0.3, s=10)
axes[0, 1].set_xlabel('Average Rooms per Household')
axes[0, 1].set_ylabel('Median House Value ($100,000s)')
axes[0, 1].set_title('Rooms vs House Value')
axes[0, 1].set_xlim(0, 15)  # Limiting x-axis to see pattern better

# HouseAge vs MedHouseVal
axes[1, 0].scatter(df['HouseAge'], df['MedHouseVal'], alpha=0.3, s=10)
axes[1, 0].set_xlabel('House Age (years)')
axes[1, 0].set_ylabel('Median House Value ($100,000s)')
axes[1, 0].set_title('Age vs House Value')

# Population vs MedHouseVal
axes[1, 1].scatter(df['Population'], df['MedHouseVal'], alpha=0.3, s=10)
axes[1, 1].set_xlabel('Block Population')
axes[1, 1].set_ylabel('Median House Value ($100,000s)')
axes[1, 1].set_title('Population vs House Value')

plt.tight_layout()
plt.show()
```

### EXAMPLE 3.2: Computing Correlations

Correlation coefficients quantify the strength of linear relationships.

```python
# Calculate correlation matrix
correlation_matrix = df.corr()

print("Correlations with MedHouseVal (target):")
target_correlations = correlation_matrix['MedHouseVal'].sort_values(ascending=False)
print(target_correlations)

# Visualize correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, fmt='.2f')
plt.title('Correlation Matrix - California Housing Dataset')
plt.tight_layout()
plt.show()
```

### EXAMPLE 3.3: Pair Plot for Quick Overview

Pair plots show all pairwise relationships at once.

```python
# Select a few key variables for pair plot (to avoid clutter)
selected_features = ['MedInc', 'HouseAge', 'AveRooms', 'MedHouseVal']
print(f"Creating pair plot for: {selected_features}")

sns.pairplot(df[selected_features], diag_kind='hist', plot_kws={'alpha': 0.3, 's': 10})
plt.suptitle('Pair Plot: Selected Features', y=1.01)
plt.tight_layout()
plt.show()
```

**Pair plots show:**
- Distributions on the diagonal
- Scatter plots for each pair of variables
